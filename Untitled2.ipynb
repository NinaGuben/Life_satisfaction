```python
import pandas as pd
import geopandas as gpd
import folium


url = 'https://gist.githubusercontent.com/Nintschi/03964299a52773bea013159206264611/raw/118d8f83aaea3be609aa6d943048274ecd96e50e/Happyness_rate.csv'
#link to data on github

url_geo = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/world-countries.json'
state_geo = gpd.read_file(url_geo)
# link to countries definitions for world map

state_data = pd.read_csv(url)
# reads the csv file from the link
```

```python
state_data = state_data [state_data.Year.eq(2018)]
state_geo
```

```python
m = folium.Map(location=[48, -102],zoom_start=2, min_zoom=2, max_bounds=True)

c = folium.Choropleth(
    geo_data=url_geo,
    name='choropleth',
    highlight = True,
    nan_fill_color = 'grey',
    data=state_data,
    columns=['Code', 'Life satisfaction in Cantril Ladder (World Happiness Report 2019)'],
    key_on='feature.id',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Life satisfaction in Cantril Ladder in 2018'
)

c.add_to(m)

folium.LayerControl().add_to(m)

state_merged = state_geo.merge (state_data, left_on='id', right_on='Code')
NIL = folium.features.GeoJson(
    state_merged,
    control=False, 
    tooltip=folium.features.GeoJsonTooltip(
        fields=['name', 'Life satisfaction in Cantril Ladder (World Happiness Report 2019)'],
        aliases=['Name', 'Life satisfaction in 2018'],
    )
)
NIL.add_to(m)

m
```
